#Basic Data Preprocesing

```
#removing last 80 rows for type error
data1 = data

#removing null spaces
print(data1["title"].isnull().sum())
print(data1["text"].isnull().sum())

data1 = data1.dropna(axis=0, subset=['text']).reset_index(drop=True)
# data1 = data1.dropna(axis=0, subset=['title']).reset_index(drop=True)
```

```
len(data1)
```

```
#website, URLs removal
pattern2 = r"(https?:\/\/) (\s)*(www\.)?(\s)* ((\w|\s)+\.)*([\w\-\s]+\/)*([\w-]+) ((\?) ?[\w\s]*=\s+[\w\%&]*)*"
pattern1 = r"((http|https)\:\/\/)?[a-zA-Z0-9\.\/\?\:@\-_=#]+\.([a-zA-Z]){2,6}([a-zA-Z0-9\.\&\/\?\:@\-_=#])*"

data1['text'] = data1['text'].str.replace(pattern1, '')
data1['title'] = data1['title'].str.replace(pattern1, '')
data1['text'] = data1['text'].str.replace(pattern2, '')
data1['title'] = data1['title'].str.replace(pattern2, '')
```

```
data1.head(5)
```

```
#removing html tags
html = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')

data1['text'] = data1['text'].str.replace(html, ' ')
data1['title'] = data1['title'].str.replace(html, ' ')
```

```
# cleaning email
email = re.compile('\S*@\S*\s')
data1['text'] = data1['text'].str.replace(email, '')
data1['title'] = data1['title'].str.replace(email, '')
```

```
#cleaning tags
tags = re.compile('@\S*\s')
data1['text'] = data1['text'].str.replace(r'@\S*\s', '')
data1['title'] = data1['title'].str.replace(tags, '')
```

```
#removing punctuations
data1['text'] = data1['text'].str.replace(r'[^\w\s]', ' ')
data1['title'] = data1['title'].str.replace(r'[^\w\s]', ' ')
```

```
#removing non ascii characters

data1['title'] = data1['title'].astype("string", copy = False)
data1['text'] = data1['text'].astype("string", copy = False)

data1['text'] = data1['text'].apply(lambda x: x.encode('ascii','ignore').decode())
# data1['title'] = data1['title'].apply(lambda x: x.encode('ascii','ignore').decode())
```

```
#removing numerical data
data1['text'] = data1['text'].str.replace(r'\d+', '')
data1['title'] = data1['title'].str.replace(r'\d+', '')
```

```
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```
#removing stopwords
pattern = re.compile(r'\b(' + r'|'.join(stopwords.words('english')) + r')\b\s*')
data1['text'] = data1['text'].str.replace(pattern, '')
data1['title'] = data1['title'].str.replace(pattern, '')
```

```
#Converting Dataset into lower-case
data1['text'] = data1['text'].str.lower()
data1['title'] = data1['title'].str.lower()
```

```
#removing extra space
data1['text'] = data1['text'].str.replace(r'\s+', ' ')
data1['title'] = data1['title'].str.replace(r'\s+', ' ')
```
